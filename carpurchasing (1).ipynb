# Which manufacturer offers the most variety when it comes to models?

```python
labels = manufacturer_counts.sort_values(ascending = False).index
manufact = manufacturer_counts.sort_values(ascending = False)

x = np.arange(len(labels))
width = 0.45

fig, ax = plt.subplots()
rects1 = ax.bar(x, manufact, width, label='Model Counts')

ax.set_ylabel('Number of Models By Manufacturer')
ax.set_title('Manufacturer')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.set_figheight(5)
fig.set_figwidth(30)

plt.show()
```

Dodge and Ford have the largest variety of models with 11 models each.

```python
fig, ax = plt.subplots(figsize=(30, 10), subplot_kw=dict(aspect="equal"))

recipe = ['Dodge', 'Ford', 'Toyota', 'Chevrolet', 'Mercedes-B', 'Mitsubishi',
       'Chrysler', 'Nissan', 'Volvo', 'Volkswagen', 'Mercury', 'Lexus',
       'Oldsmobile', 'Pontiac', 'Honda', 'Cadillac', 'Saturn', 'Buick',
       'Plymouth', 'Acura', 'Hyundai', 'Jeep', 'Lincoln', 'BMW', 'Audi',
       'Porsche', 'Saab', 'Subaru', 'Infiniti', 'Jaguar']

data = manufacturer_counts.sort_values(ascending = False)

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.6), startangle=-100)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Models By Manufacturer")

plt.show()
```

# Which car has the highest fuel efficiency and which car has the lowest fuel efficiency? How about price?

```python
def max_min_val(col):

    first = df[col].idxmax()
    first_obs = pd.DataFrame(df.loc[first])
    
    last = df[col].idxmin()
    last_obs = pd.DataFrame(df.loc[last])
    
    min_max_obs = pd.concat([first_obs, last_obs], axis=1)
    
    return min_max_obs 
```

```python
max_min_val('Fuel_efficiency')
```

```python
max_min_val('Price_in_thousands')
```

The min max function allows you to pull up information on any vehicle given any feature.

We can see that the Chevy Metro has the best fuel effiency while the Cadillac Escalade has the worst.
